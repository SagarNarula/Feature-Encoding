### Count or Frequency Encoding : Basically it applies to the categorical variables which has large number of labels 

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('mercedesbenz.csv',usecols=['X1','X2'])
data.head()
```

```python
for i in data.columns:
    print(i ,':',data[i].nunique())
```

```python
temp=dict(data['X2'].value_counts())
temp
```

```python
freq_encode=[]
for i in data.X2:
    if i in temp.keys():
        freq_encode.append(temp[i])
data['X2']=freq_encode
```

```python
freq_encode
```

```python
data.head()
```

Advantages:
1.Easy to Implement.
2.Does not increase the feature dimensional space

Disadvantage:
1.If some of the labels have the same count,then they will be replaced with the same count,and they will loose some information.

### One Hot Encoding for Multiple Categories:

```python
## Using the same data 
## when you have a column with multiple categories:
data=pd.read_csv('mercedesbenz.csv',usecols=['X1','X2'])
data.head()
```

```python
for i in data.columns:
    print(i ,':',data[i].nunique())
```

```python
# Lets find the top 10 most frequent labels for the variable X2.
data['X2'].value_counts().head(10)
```

```python
# Lets make a list with the top10 categories :
top_10=[x for x in data['X2'].value_counts().head(10).index]
top_10
```

```python
##now we will make 10 binary variables:
for label in top_10:
    data[label]=np.where(data['X2']==label,1,0)
data.head(10)
```

```python
### Creating the function:
def one_hot_encoding(data,variable,top_10):
    for label in top_10:
        data[variable+"_"+label]=np.where(data[variable]==label,1,0)
        
top_10=[x for x in data['X1'].value_counts().head(10).index]
one_hot_encoding(data,'X1',top_10)
        
```

```python
data.head()
```

Advantages:
1.Easy to implement.
2.Does not requires hours of variable exploration

Disadvantages:
1.Does not add any information that may make the variable more predictive.

```python
## Target Mean Encoding:
import seaborn as sns
data=sns.load_dataset('planets')
data.head()
```

```python
## Here our orbital period is our independent variable:So here we will calculate the mean with respect to our target columns:
```

```python
mean_encode=dict(data.groupby('method')['orbital_period'].mean())
```

```python
mean_encode
```

```python
tmean_encode=[]
for i in data['method']:
    if i in mean_encode.keys():
        tmean_encode.append(mean_encode[i])
tmean_encode
        
```

```python
data['method']=tmean_encode
data.head()
```
